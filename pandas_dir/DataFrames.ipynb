```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import csv

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# Read individual CSV files 
Winter_file = "top10parks_winter.csv"
Spring_file = "top10parks_spring.csv"
Summer_file = "top10parks_summer.csv"
Fall_file = "top10parks_fall.csv" 
StatePopulation_file = "PopulationPerState.csv"
AgeRange_file = "Age_Percentage.csv"
visitors_file = "VisitationPerSeason.csv"
meanIncome_file = "MeanIncomeByAgeAndRace.csv"
medIncome_file = "MedianIncomeByAgeAndRace.csv"
```
