```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import csv
import requests
import urllib.request, json
%matplotlib inline
import matplotlib.pyplot as plt
import folium


# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# Read individual CSV files 
national_parks_usa_WINTER_to_load = "C:/Users/ali89/Desktop/JADBY/csv_dir_correct/top10parks_by_Winter_1.csv"
national_parks_usa_SPRING_to_load = "C:/Users/ali89/Desktop/JADBY/csv_dir_correct/top10parks_by-season_spring_1.csv"
national_parks_usa_SUMMER_to_load = "C:/Users/ali89/Desktop/JADBY/csv_dir_correct/top10parks_by-season_summer_1.csv"
national_parks_usa_FALL_to_load = "C:/Users/ali89/Desktop/JADBY/csv_dir_correct/top10parks_by_Fall_1.csv" 


# Create individual Data Frame
national_parks_usa_WINTER = pd.read_csv(national_parks_usa_WINTER_to_load,keep_default_na=False, na_values=[""])
national_parks_usa_SPRING = pd.read_csv(national_parks_usa_SPRING_to_load,keep_default_na=False, na_values=[""])
national_parks_usa_SUMMER = pd.read_csv(national_parks_usa_SUMMER_to_load,keep_default_na=False, na_values=[""])
national_parks_usa_FALL = pd.read_csv(national_parks_usa_FALL_to_load,keep_default_na=False, na_values=[""])

# This is a very important step, DO NOT SKIP .
Second_df_to_append = national_parks_usa_SPRING.reset_index()
Second_df_set2_append = national_parks_usa_FALL.reset_index()
```

```python
# concatinate two Data frames . DO vertical Stacking . Make sure the columns are same . 
Winter_Spring_Parks_collections_df = pd.concat([national_parks_usa_WINTER,Second_df_to_append ], axis=0)
Summer_Fall_Parks_collections_df = pd.concat([national_parks_usa_SUMMER,Second_df_set2_append ], axis=0) 
```

```python
# Load coordinate data of each national park.
parks_coord = pd.read_csv("C:/Users/ali89/Desktop/Park_Coords.csv")

# Display the DataFrame generated from loaded data
parks_coord.head()
```

```python
# Merging data
parks_df = pd.merge(Summer_Fall_Parks_collections_df, parks_coord, on='Park')

# Display the DataFrame generated from merged data
parks_df
```

```python
# Select the first occurrence of each station id.
parks_df = parks_df.groupby("Park").first()

# Display the new DataFrame
parks_df.head()
```

```python
# Select Start Station Name, Latitude, and Longitude as columns in DataFrame
parks_df = parks_df.loc[:, ["Rating 1,3,5,4","Latitude","Longitude","Top Activity"]]
parks_df
```

```python
# Generate a map from one park location
parks_map = folium.Map(location=[61.4182147,-142.6028439],
                        zoom_start=2.9,
                        width='100%')

# Display map
parks_map
```

```python
# Extract the location, name, and rating information for each park
for index, row in parks_df.iterrows():
    rating = row['Rating 1,3,5,4']
    lat = row['Latitude']
    lon = row['Longitude']
    activity = row['Top Activity']
    name = index
    
    if rating < 3.0:
        index = folium.Icon(color='red')
    else:
        index = folium.Icon(color='green')
        
    ##
    popup = (
        "{name}<br>"
        "Rating: {rating}<br>"
        "Activities: {activity}<br>"
               ).format(name=str(name), rating=str(round(rating,2)), activity=str(activity))
    ##

    #popup = folium.Popup(labels[row], parse_html=True)
    folium.Marker([lat, lon], popup=popup, icon = index).add_to(parks_map)    
    
# Create the html file with the map:
parks_map.save(outfile='parks_map.html')

# Display map
parks_map

#print(name)
```
